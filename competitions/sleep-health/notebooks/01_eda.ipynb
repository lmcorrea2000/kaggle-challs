Dataset: [Kaggle - Sleep Health and Lifestyle](https://www.kaggle.com/datasets/uom190346a/sleep-health-and-lifestyle-dataset)

**Perguntas que queremos responder:**
- Quais fatores mais influenciam a qualidade do sono?
- Existe relação entre atividade física e distúrbios do sono?
- Como ocupação e estresse se relacionam com a saúde do sono?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 5)

df = pd.read_csv('../data/Sleep_health_and_lifestyle_dataset.csv')
print(f'Shape: {df.shape}')
df.head()
```

## 1. Visão Geral

```python
df.info()
```

```python
df.describe()
```

```python
# Valores nulos
null_pct = df.isnull().sum() / len(df) * 100
print('Valores nulos (%):')
print(null_pct[null_pct > 0])
```

## 2. Distribuição das Variáveis Numéricas

```python
num_cols = ['Age', 'Sleep Duration', 'Quality of Sleep', 'Physical Activity Level',
            'Stress Level', 'Heart Rate', 'Daily Steps']

fig, axes = plt.subplots(2, 4, figsize=(16, 8))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    axes[i].hist(df[col], bins=20, color='steelblue', edgecolor='white')
    axes[i].set_title(col)

axes[-1].set_visible(False)
plt.tight_layout()
plt.show()
```

## 3. Variáveis Categóricas

```python
cat_cols = ['Gender', 'Occupation', 'BMI Category', 'Sleep Disorder']

fig, axes = plt.subplots(1, 4, figsize=(18, 5))

for i, col in enumerate(cat_cols):
    counts = df[col].value_counts()
    axes[i].bar(counts.index, counts.values, color='steelblue')
    axes[i].set_title(col)
    axes[i].tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()
```

## 4. Correlação entre Variáveis Numéricas

```python
corr = df[num_cols].corr()

plt.figure(figsize=(10, 7))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Correlação entre Variáveis Numéricas')
plt.show()
```

## 5. Qualidade do Sono por Grupo

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# Por gênero
df.groupby('Gender')['Quality of Sleep'].mean().plot(kind='bar', ax=axes[0], color='steelblue')
axes[0].set_title('Qualidade do Sono por Gênero')
axes[0].set_ylabel('Média')
axes[0].tick_params(axis='x', rotation=0)

# Por BMI
df.groupby('BMI Category')['Quality of Sleep'].mean().plot(kind='bar', ax=axes[1], color='steelblue')
axes[1].set_title('Qualidade do Sono por BMI')
axes[1].tick_params(axis='x', rotation=30)

# Por distúrbio do sono
df.groupby('Sleep Disorder')['Quality of Sleep'].mean().plot(kind='bar', ax=axes[2], color='steelblue')
axes[2].set_title('Qualidade do Sono por Distúrbio')
axes[2].tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()
```

## 6. Atividade Física vs Qualidade do Sono

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].scatter(df['Physical Activity Level'], df['Quality of Sleep'], alpha=0.5, color='steelblue')
axes[0].set_xlabel('Nível de Atividade Física')
axes[0].set_ylabel('Qualidade do Sono')
axes[0].set_title('Atividade Física vs Qualidade do Sono')

axes[1].scatter(df['Stress Level'], df['Quality of Sleep'], alpha=0.5, color='tomato')
axes[1].set_xlabel('Nível de Estresse')
axes[1].set_ylabel('Qualidade do Sono')
axes[1].set_title('Estresse vs Qualidade do Sono')

plt.tight_layout()
plt.show()
```

## 7. Qualidade do Sono por Ocupação

```python
occ_sleep = df.groupby('Occupation')['Quality of Sleep'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 5))
occ_sleep.plot(kind='bar', color='steelblue')
plt.title('Qualidade Média do Sono por Ocupação')
plt.ylabel('Qualidade do Sono (média)')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```
